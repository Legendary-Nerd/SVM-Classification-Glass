# SVM
A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

## Data Acquisition

```python
gd = pd.read_csv('/content/drive/MyDrive/glass.csv')
gd.head(5)
```

```python
gd.describe()
```

```python
gd.shape
```

## Data Preparation

```python
print("Ba:",(gd['Ba']==0).sum()/len(gd))
print("Fe:",(gd['Fe']==0).sum()/len(gd))
```

Here we see 'Ba'  and 'Fe' mostly 0 <br>
hence drop Ba and Fe columns

```python
gd.drop('Ba',axis=1,inplace=True)
gd.drop('Fe', axis=1, inplace=True)

gd
```

```python
X = gd.iloc[:,:-1]
y = gd.iloc[:,-1]

X_train , X_test , y_train, y_test = train_test_split(X, y ,train_size=0.2 , random_state=42)

X_train , y_train
```

## Model Selection and Training

```python
model = SVC(kernel='linear' ,C=0.8)

model.fit(X_train,y_train)
```

Model Evaluation

```python
y_predict = model.predict(X_test)

print(accuracy_score(y_test,y_predict))
```

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```
